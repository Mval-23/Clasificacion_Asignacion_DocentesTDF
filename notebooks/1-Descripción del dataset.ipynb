## Análisis exploratorio visual

```python
# 1. Distribución de la variable objetivo
sns.countplot(data=dat_excel, x='caracter', palette='Set2')
plt.title('Distribución de Categorías Docentes')
plt.xlabel('Clasificación')
plt.ylabel('Cantidad')
plt.show()
```

```python
# 2. Títulos más frecuentes
top_titulos = dat_excel['titulo'].value_counts().nlargest(10)
top_titulos.plot(kind='barh', color='skyblue')
plt.title('Top 10 Títulos más frecuentes')
plt.xlabel('Cantidad')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 3. Espacios curriculares más asignados
top_espacios = dat_excel['desc_espacio'].value_counts().nlargest(10)
top_espacios.plot(kind='bar', color='salmon')
plt.title('Top 10 Espacios Curriculares')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Relación entre clasificación y espacio curricular

sns.countplot(data=dat_excel, x='caracter', hue='desc_espacio', order=['Docente', 'Habilitante', 'Supletorio'])
plt.title('Distribución de clasificación por espacio curricular')
plt.xticks(rotation=45)
plt.legend(title='Espacio', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
